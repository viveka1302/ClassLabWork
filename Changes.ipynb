```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread(r"Suit pic.jpg", cv.IMREAD_GRAYSCALE)
plt.imshow(img,cmap="gray")
```

```python

    kernelX = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
    kernelY = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
    prewittX = cv.filter2D(img,-1,kernelX)
    prewittY = cv.filter2D(img,-1,kernelY)
    prewitt = prewittX + prewittY
    print('Prewitt Edge Detection :')
    plt.imshow(prewitt,cmap='gray')
```

```python
    sobelX = cv.Sobel(img,cv.CV_8U,1,0,ksize=5)
    sobelY = cv.Sobel(img,cv.CV_8U,0,1,ksize=5)
    sobel = sobelX + sobelY
    print('Sobel Edge Detection :')
    plt.imshow(sobel,cmap='gray')
    
```

```python
    kernelX = np.array([[-3,0,3],[-10,0,10],[-3,0,3]])
    kernelY = np.array([[-3,-10,-3],[0,0,0],[3,10,3]])
    scharrX = cv.filter2D(img,-1,kernelX)
    scharrY = cv.filter2D(img,-1,kernelY)
    scharr = scharrX + scharrY
    print('Scharr Edge Detection :')
    plt.imshow(scharr,cmap='gray')
```
